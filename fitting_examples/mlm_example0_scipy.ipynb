```python
import numpy as np
import matplotlib.pylab as plt
import math as math
from scipy.optimize import fmin
%matplotlib  notebook

#generate and plot fake data
Nstudents = 200
fake_mean = 80
fake_width = 5
scores = np.random.normal(fake_mean,fake_width,Nstudents)
plt.figure()
plt.hist(scores,bins=25)#,range=(60,100))

def Gaussian(x,mean,width):

        return (1.0/(width*np.sqrt(2*np.pi)))*(np.exp(-(x-mean)**2/(2*(width**2))))

#original values to start at
lower_mean=70
lower_width=5

#the function being minimized
def probability(params,x):
    mean = params[0]
    width = params[1]
    return -np.log(1.0/(width*np.sqrt(2*np.pi))*np.exp(-(x-mean)**2/(2*(width**2)))).sum()

optimal_params=fmin(probability, [lower_mean,lower_width],args=(scores,))
print optimal_params

#extracting the best mean and width
best_mean=optimal_params[0]
best_width=optimal_params[1]

#plotting the Gaussian
x = np.linspace(60,100,1000)
y = Gaussian(x,best_mean,best_width)*Nstudents
plt.plot(x,y, linewidth=4)
plt.show()


```
