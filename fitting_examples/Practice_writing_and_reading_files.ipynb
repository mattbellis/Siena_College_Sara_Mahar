```python
import matplotlib.pylab as plt
%matplotlib notebook
```

```python
outfilename = "test.dat"
outfile = open(outfilename,'w')

output = ""
for i in range(10):
    output = "%d %d %f\n" % (i,i+2,float(i*24))
    outfile.write(output)

outfile.close()

```

```python
# Now lets read it in

import numpy as np

infilename = outfilename

values = np.loadtxt(infilename,unpack=True,delimiter=' ',dtype=float)
```
